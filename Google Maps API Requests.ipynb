```python
# Functions
import pandas as pd
import requests

def get_coordinates(df):
    
    # Rename the columns for clarity
    df.columns = ['Longitude', 'Latitude']
    
    # Convert the DataFrame to a list of tuples
    coordinates_list = list(df.itertuples(index=False, name = None))
    
    return coordinates_list

def get_walking_distance(api_key, origin, destination):
    url = "https://maps.googleapis.com/maps/api/distancematrix/json"
    params = {
        'origins': f'{origin[1]},{origin[0]}',
        'destinations': f'{destination[1]},{destination[0]}',
        'mode': 'walking',
        'key': api_key
    }
    
    try:
        response = requests.get(url, params=params)
        response.raise_for_status()
        result = response.json()
        
        try:
            distance = result['rows'][0]['elements'][0]['distance']['value']
            duration = result['rows'][0]['elements'][0]['duration']['value']
            return distance, duration
        except (IndexError, KeyError):
            print("Error parsing response for origins:", origin, "and destinations:", destination)
            return None, None

    except requests.exceptions.RequestException as e:
        print(f"Request error: {e}")
        return None, None
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
        return None, None
    
def get_driving_distance(api_key, origin, destination):
    url = "https://maps.googleapis.com/maps/api/distancematrix/json"
    params = {
        'origins': f'{origin[1]},{origin[0]}',
        'destinations': f'{destination[1]},{destination[0]}',
        'mode': 'driving',
        'key': api_key
    }
    
    try:
        response = requests.get(url, params=params)
        response.raise_for_status()
        result = response.json()
        
        try:
            distance = result['rows'][0]['elements'][0]['distance']['value']
            duration = result['rows'][0]['elements'][0]['duration']['value']
            return distance, duration
        except (IndexError, KeyError):
            print("Error parsing response for origins:", origin, "and destinations:", destination)
            return None, None

    except requests.exceptions.RequestException as e:
        print(f"Request error: {e}")
        return None, None
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
        return None, None
```

```python
# ROTTERDAM: Combine all coordinate points into one with the following order: DPs, PUs, Shelters

# Load the CSV file with coordinates
shelter_coordinates = pd.read_csv("path")
ci_coordinates = pd.read_csv("path")
ps_coordinates = pd.read_csv("path")
ts_coordinates = pd.read_csv("path")
r_coordinates = pd.read_csv("path")
fl_coordinates = pd.read_csv("path")

# Combine hospitals and education locations into one DataFrame
all_points = pd.concat([shelter_coordinates, ci_coordinates, ps_coordinates, ts_coordinates, r_coordinates, fl_coordinates], ignore_index=True)

# Save the combined DataFrame to a new CSV file
all_points_file_path = "path"
all_points.to_csv(all_points_file_path, index = False, header = False)

coordinates = get_coordinates(all_points)
```
